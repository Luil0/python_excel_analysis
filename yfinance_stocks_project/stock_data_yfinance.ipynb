```python
import yfinance as yf 
import pandas as pd 
from datetime import datetime, timedelta
import numpy as np
import matplotlib.pyplot as plt
```

```python
class GetStockData:
    
    def __init__(self, tickers, start=datetime.today()-timedelta(days=365), end=datetime.today()):
        self.tickers = tickers
        self.stocks_data = yf.download(tickers, start, end)
        
    # gets the data for a specific stock that is in our stocks_data.
    def get_symbol(self, symbol):
        stocks_df = self.stocks_data.loc[:, self.stocks_data.columns.get_level_values(0).isin(["Date"]) | self.stocks_data.columns.get_level_values(1).isin([symbol])]
        return stocks_df
    
    # Get monthly data for a stock or the stocks that were in the ticker
    def monthly_data_ct(self, symbol=None):
        by_month = self.get_symbol(symbol).groupby(pd.Grouper(level="Date",freq="M"))
        full_data_by_month = self.stocks_data.groupby(pd.Grouper(level="Date", freq="M"))
        if symbol == None:
            ct = full_data_by_month
            
        else:
            ct = by_month
        
        # getting central tendency for our dataframe
        result = {"Mean":ct.mean(),
        "Max":ct.max(), "Min":ct.min()}
        result["Mean"].rename_axis("Month", inplace=True)
        result["Mean"].index = result["Mean"].index.strftime('%b-%Y')
        return result
        
        
        

tick = ["AAPL", "MSFT", "TSLA"]



df = GetStockData(tick).monthly_data_ct()["Mean"]

df.columns = [col[0]+" "+col[1] for col in df.columns]
df.rename(columns = {"Date":"Month"})

# df.rename_axis("Month")
# df.index = df.index.strftime("%B")
df
```

```python
df.to_csv("yfinance_mean_data_python.csv")
```
